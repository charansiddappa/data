```python
import pandas as pd

# Create two dataframes for the salary of five employees
data1 = {
    'Name': ['John', 'Alice', 'Bob', 'Eve', 'Charlie'],
    'Salary': [50000, 60000, 55000, 70000, 65000]
}

data2 = {
    'Name': ['David', 'Ella', 'Frank', 'Grace', 'Hannah'],
    'Salary': [72000, 48000, 51000, 78000, 62000]
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

# Display both dataframes
print("Original DataFrame 1:")
print(df1)
print("\nOriginal DataFrame 2:")
print(df2)

# Add 5000 as bonus to both dataframes
df1['Salary'] = df1['Salary'] + 5000
df2['Salary'] = df2['Salary'] + 5000

# Display the updated dataframes
print("\nUpdated DataFrame 1 with Bonus:")
print(df1)
print("\nUpdated DataFrame 2 with Bonus:")
print(df2)
```

```python
import pandas as pd

# Create the dataframe CORONA
data = {
    'ID': [100, 110, 120, 130],
    'State': ['Delhi', 'Mumbai', 'Chennai', 'Surat'],
    'Cases': [3000, 4000, 5000, 4500]
}
corona_df = pd.DataFrame(data)

# Set the ID column as the index
corona_df.set_index('ID', inplace=True)

# Display the initial dataframe
print("Initial DataFrame:")
print(corona_df)

# (a) Add a new column "Recovery" using the series method
recovery_series = pd.Series([2500, 3500, 4500, 4000], index=[100, 110, 120, 130])
corona_df['Recovery'] = recovery_series

# (b) Add a new column "Deaths" using the assign() method
corona_df = corona_df.assign(Deaths=[100, 200, 300, 250])

# Display the updated dataframe
print("\nUpdated DataFrame with Recovery and Deaths columns:")
print(corona_df)
```

```python
import pandas as pd

# Sample DataFrame
df = pd.DataFrame({
    'State': ['State1', 'State2'],
    'Cases': [1000, 1500],
    'Deaths': [50, 75]
})

print("Initial DataFrame:")
print(df)

# (c) Adding a new row using loc
df.loc[len(df)] = ['State3', 2000, 100]  # Assuming values for 'State3'

print("\nDataFrame after adding a new row:")
print(df)

# (d) Adding a new column "Percentage" as the fourth column
df.insert(3, 'Percentage', [5.0, 7.5, 10.0])  # Assuming percentage values

print("\nDataFrame after adding the 'Percentage' column:")
print(df)

# (e) Deleting the column "Percentage" using del
del df['Percentage']

print("\nDataFrame after deleting the 'Percentage' column:")
print(df)

# (f) Deleting the column "Deaths" using pop()
deaths = df.pop('Deaths')

print("\nDataFrame after deleting the 'Deaths' column:")
print(df)
print("\nRemoved 'Deaths' column data:")
print(deaths)

# (g) Inserting a new row at the 1st position using iloc
new_row = pd.DataFrame([['NewState', 500, 10]], columns=df.columns)
df = pd.concat([new_row, df]).reset_index(drop=True)

print("\nDataFrame after inserting a new row at the 1st position:")
print(df)

# (h) Temporarily deleting columns "Cases" and "State"
temporary_df = df.drop(columns=['Cases', 'State'])

print("\nTemporary DataFrame after deleting 'Cases' and 'State' columns:")
print(temporary_df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Create the data frame
data = {
    'Ename': ['Roshan', 'Amar', 'Ashwini', 'Lohith', 'Mohan', 'Pramod'],
    'Type': ['Regular', 'Adhoc', 'Regular', 'Adhoc', 'Contract', 'Regular'],
    'Department': ['CS', 'CS', 'EC', 'EC', 'CS', 'EC'],
    'Experience': [10, 20, 5, 14, 9, 8],
    'Salary': [50000, 15000, 30000, 15000, 10000, 40000]
}

df = pd.DataFrame(data)

# Pivot table for average salary of each employee for each department
pivot_avg_salary = df.pivot_table(values='Salary', index='Department', aggfunc='mean')
print("Average Salary for Each Department:")
print(pivot_avg_salary)

# Pivot table for sum and mean of the salaries of each type of employee and the number of employees of each type
pivot_type_salary = df.pivot_table(values='Salary', index='Type', aggfunc=['sum', 'mean', 'count'])
print("\nSum, Mean, and Count of Salaries for Each Type:")
print(pivot_type_salary)

# Visualizations
# Average salary per department
plt.figure(figsize=(10, 5))
sns.barplot(x=pivot_avg_salary.index, y='Salary', data=pivot_avg_salary.reset_index())
plt.title('Average Salary per Department')
plt.xlabel('Department')
plt.ylabel('Average Salary')
plt.savefig('/mnt/data/average_salary_per_department.png')
plt.show()

# Sum of salaries per type
plt.figure(figsize=(10, 5))
sns.barplot(x=pivot_type_salary.index, y=('sum', 'Salary'), data=pivot_type_salary.reset_index())
plt.title('Sum of Salaries per Type')
plt.xlabel('Type')
plt.ylabel('Sum of Salaries')
plt.savefig('/mnt/data/sum_of_salaries_per_type.png')
plt.show()

# Mean of salaries per type
plt.figure(figsize=(10, 5))
sns.barplot(x=pivot_type_salary.index, y=('mean', 'Salary'), data=pivot_type_salary.reset_index())
plt.title('Mean of Salaries per Type')
plt.xlabel('Type')
plt.ylabel('Mean of Salaries')
plt.savefig('/mnt/data/mean_of_salaries_per_type.png')
plt.show()

# Count of employees per type
plt.figure(figsize=(10, 5))
sns.barplot(x=pivot_type_salary.index, y=('count', 'Salary'), data=pivot_type_salary.reset_index())
plt.title('Number of Employees per Type')
plt.xlabel('Type')
plt.ylabel('Number of Employees')
plt.savefig('/mnt/data/number_of_employees_per_type.png')
plt.show()

# Git commands
repo_name = "employee_data_analysis"
os.system(f"git init {repo_name}")
os.chdir(repo_name)
os.system("git add .")
os.system('git commit -m "Initial commit with data analysis"')
os.system(f"git remote add origin <YOUR_GITHUB_REPO_URL>")
os.system("git push -u origin master")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Create the DataFrame
data = {
    'Name': ['charan','siddappa','hemanth','saad'],
    'Maths': [85, 75, 96, 87],
    'Science': [95, 50, 65, 80],
    'Social': [50, 55, 56, 60],
    'English': [60, 85, 89, 65]
}

df = pd.DataFrame(data)

# Step 2: Create a Pivot Table
pivot_table = df.melt(id_vars='Name', var_name='Subject', value_name='Score')

# Step 3: Add Rows and Columns
new_student = pd.DataFrame({'Name': ['NewStudent'], 'Maths': [70], 'Science': [80], 'Social': [75], 'English': [90]})
df = pd.concat([df, new_student], ignore_index=True)
df['Total Score'] = df[['Maths', 'Science', 'Social', 'English']].sum(axis=1)

# Step 4: Sort the Data by Specific Subject
sorted_df = df[df['Maths'] > 80].sort_values(by='Maths', ascending=False)

# Step 5: Visualize with Appropriate Plots
plt.figure(figsize=(10, 6))
plt.bar(df['Name'], df['Total Score'], color=['blue', 'orange', 'green', 'red', 'purple'])
plt.xlabel('Name of Students',fontsize=14)
plt.ylabel('Total Score')
plt.title('Total Score of Each Student')
plt.show()
```
