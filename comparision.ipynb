```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some data
data1 = np.random.randn(100)
data2 = np.random.randn(100) + 2  # Shift the second dataset for contrast

# Create side-by-side histograms
plt.hist(data1, bins=15, alpha=0.5, label='Data 1')
plt.hist(data2, bins=15, alpha=0.5, label='Data 2')

# Add titles and labels
plt.title('Side by Side Histograms')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Generate some example data
np.random.seed(10)
data = pd.DataFrame({
    'Category': np.repeat(['A', 'B', 'C'], 50),
    'Value': np.concatenate([np.random.randn(50) + i for i in range(3)])
})

# Create and display the swarm plot
sns.swarmplot(x='Category', y='Value', data=data)
plt.title('Swarm Plot')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Generate some example data
np.random.seed(10)
data = pd.DataFrame({
    'Category': np.repeat(['A', 'B', 'C'], 50),
    'Value': np.concatenate([np.random.randn(50) + i for i in range(3)])
})

# Create and display the beeswarm (swarm) plot
sns.swarmplot(x='Category', y='Value', data=data)
plt.title('Beeswarm Plot')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
sizes = [15, 30, 45, 10]
labels = ['A', 'B', 'C', 'D']
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']

# Create a pie chart
plt.pie(sizes, labels=labels, colors=colors, wedgeprops=dict(width=0.3))

# Draw a circle at the center of the pie to make it a donut
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Equal aspect ratio ensures that pie is drawn as a circle
plt.title('Donut Plot')
plt.axis('equal')  

# Show the plot
plt.show()
```
